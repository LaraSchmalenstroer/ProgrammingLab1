### Ex. 1 Srinivasa Ramanujan calculates $\pi$

#### a) (2 pts) Write a function called estimate_pi that uses this formula to compute and return an estimate of $\pi$.  It should use a while loop to compute terms of the summation until the last term is smaller than 1e-15 (which is Python notation for 10−15). (‘Term’refers to the expression after the summation sign, not the summation itself.) You can check the result by comparing it to math.pi.  The error should be less than10−15.

```python
def estimate_pi():
    """The function calculates an estimate of pi using Ramanujan's formula"""
    from math import factorial
    factor = (2*(2**0.5))/9801
    k = 0
    summation=0
    sum_h=1
    #in each iteration of the while loop the current and previous sums are compared
    #if the difference between those two sums is smaller than 1e-15, the while loop terminates
    while abs(sum_h-summation) >= 1e-15:
        sum_h=summation
        summation += (factorial(4*k)*(1103+26390*k))/((factorial(k)**4)*396**(4*k))
        k+=1
    result=1/(factor*summation)
    return result
```

#### Happy numbers are defined by the following process: Start with a positive number.  Replacethe number with the sum of the squares of its digits and repeat until you reach the number1 or the process enters a loop not involving the number 1. #### (a) (1 pt) Write a function isHappy(n) that checks whether a number is happy or un-happy.  It should return true if the the number is happy and false otherwise.

##### The first function checks this recursively, the second function uses a while loop

```python
def isHappy_recursive(n):
    """The function determines recursivey whether a number is happy or not. As input, only 
    positive integers are allowed."""
    assert type(n)==int, "please enter an integer"
    assert n>0, "please enter a positive integer"
    #every unhappy number passes a certain circle of numbers that are contained in the unhappy list
    unhappy=[4,16,37,58,89,145,42,20]
    string_n=str(n)
    digits=[int(char) for char in string_n]
    sum_squares=sum(digit**2 for digit in digits)
    if sum_squares in unhappy:
        return False
    elif sum_squares==1:
        return True
    else:
        return isHappy(sum_squares)
```

```python
isHappy_recursive(10)
```

```python
def isHappy(n):
    """The function determines the happiness of a number using a while loop. Only positive 
    integers are allowed as input."""
    assert type(n)==int, "please enter an integer"
    assert n>0, "please enter a positive integer"
    unhappy=[4,16,37,58,89,145,42,20]
    string_n=str(n)
    digits=[int(char) for char in string_n]
    sum_squares=sum(digit**2 for digit in digits)
    while sum_squares!=1:
        digits=[int(char) for char in str(sum_squares)]
        sum_squares=sum(digit**2 for digit in digits)
        if sum_squares in unhappy:
            return False
    return True
```

### Ex.  3 (4 pts) The Birthday Paradox
#### (a) (1 pt) Write a function called has_duplicatesthat takes a list and returns True ift here is any element that appears more than once.  It should not modify the original list.


```python
def has_duplicates(input_list):
    """The function detects if a given input list contains duplicates. It returns True if at least
    one element appears more than once, else False."""
    for i in range(len(input_list)):
        for j in range(i+1,len(input_list)):
            if input_list[i]==input_list[j]:
                return True
    return False
```

#### (b)  If there are n = 27 students in your class, what are the chances that two of you have the  same  birthday  (day/month)?   You  can  estimate  this  probability  by  generatin grandom samples of n birthdays and checking for matches.  (For simplicity, assume thatevery year has 365 days and the probability to be born on any day is the same.) Hint: You can generate random birthdays with the randint function in the random module.
##### (1 pts) Estimate the probability on the basis generating 10 000 trials of n= 27 birthdays and determine the fraction of trials, where at least two persons share a birthday.

```python
def birthday_paradox(persons_n,trials_n):
    import random
    count_duplicates=0
    for i in range(trials_n):
        birthdays=[random.randint(1,365) for i in range(persons_n)]
        if has_duplicates(birthdays):
            count_duplicates+=1
    ratio=count_duplicates/trials_n
    return ratio
```

```python
test_ratio=birthday_paradox(27,10000)
test_ratio
```

##### (1 pt) Modify your approach to estimate the probability that at least three people have a non-unique birthday in their class. Hint: Read carefully!  In a class with two  pairs  of  people  where  each  pair  shares  a  single  birthday  there  are  also  ”at least three” people having a non-unique birthday.

```python
def birthday_paradox_mod(persons_n,trials_n):
    """The function calculates the probability that out of a given number of persons at least 
    3 have a non-unique birthday by calculating random sets of birthdays a given number 
    (trials_n) of times)"""
    import random
    count=0
    for i in range(trials_n):
        birthdays=[random.randint(1,365) for i in range(persons_n)]
        duplicates=[0 for i in range(max(birthdays)+1)]
        for j in range(len(birthdays)):
            duplicates[birthdays[j-1]]+=1
        if max(duplicates)>2 or duplicates.count(2)>=2:
            count+=1
    return count/trials_n
```

There are two different possibilities of at least three persons having non-unique birthdays:
+ there are at least two couples sharing a birthday
+ there are at least three people sharing a birthday

To calculate the probability, the function calculates a random set of birthdays. Then a second list with the length of the maximum value of birthdays, increased by 1, is initialized with 0 at every position. Then the for loop iterates over the birthday list. The value is used to index the duplicate list by 1. 
The conditions for at least three persons with a non-unique birthday is met if there are at least 2 times values greater than 2 in duplicates (duplicates.count(2)>=2) or once a value of at least 3 (max(duplicates)>2)

### Ex. 4 Making triangles 

#### Imagine you are given a stick of length 1 (meter) and you break the stick randomly at two points leaving you with 3 smaller sticks of random length.  How likely is it that the three sticks can be combined to form 3 sides of a triangle?  (See exercise 5.3 of ThinkPython.) Write Python code to simulate 1 000 000 trials splitting the stick randomly into 3 pieces and estimate the probability that a triangle can be formed from the three pieces.

```python
def check_triangle(length):
    """The function checks the possibility of creating a triangle out of parts of a stick that\
     is randomly split into three pieces."""
    from random import random, uniform
    cutting_sites=[uniform(0,length),uniform(0,length)]
    cuts=sorted(cutting_sites)
    a=cuts[0]
    c=length-cuts[1]
    b=length-(a+c)
    pieces=sorted([a,b,c])
    #a triangle is only valid if the sum of the shorter sides is larger than the longest side
    if pieces[0]+pieces[1]>pieces[2]:
        return True
    return False
```

```python
check_triangle(1)
```

```python
counter_triangles=0
for i in range(1000000):
    if check_triangle(1)==True:
        counter_triangles+=1

prob = counter_triangles/1000000
prob
```

### Ex. 5 Estimating $\pi$

```python
def estimate_pi_MC(n):
    """The function uses Monte-Carlo-Integration to estimate the value of pi. The input parameter\
    n indicates how many random points are picked from the intervall -1 to 1."""
    import random
    sum_in_circle=0
    for i in range(n):
        x_val=random.uniform(-1,1)
        y_val=random.uniform(-1,1)
        if ((x_val**2+y_val**2)<=1):
            sum_in_circle+=1
    ratio=sum_in_circle/n
    estimation=4*ratio
    return estimation
        
#the x and y values are randomly picked from the interval from -1 to 1. 
#The point is in the circle if the condition x^2 + y^2 <= 1 is met
```

```python
dict_estimation={}
list_test=[10,100,1000,10000,100000,1000000]
for i in range(len(list_test)):
    dict_estimation[list_test[i]]=estimate_pi_MC(list_test[i])
```

```python
print(dict_estimation)
```

The approximation of $\pi$ improves with an increasing number of values that are randomly picked from the given square. In the first step (10), the estimation is not correct. In the steps with 100 calculations, the first digit is correct. With 100, 1000 and 10000 calculations, two digits of $\pi$ are correct. The calculation is correct up to the 3th digit when calculating 100000 and 1000000 times.

#### (a) (2 pts) Write a program that reads a word list from the file words.txt and prints all the sets of words that are anagrams.  Limit your output to words having at least 6 anagrams (including itself). Here is an example of what the output might look like:

#### ['deltas', 'desalt', 'lasted', 'salted', 'slated', 'staled']

#### ['retainers', 'ternaries']

#### ['generating', 'greatening']

#### ['resmelts', 'smelters', 'termless']

```python
with open("words.txt", 'r') as file:
    content=file.read()
list_words=content.split('\n')
```

```python
def anagram_finder(list_words):
    """The function finds all sets of anagrams in a given list of words."""
    dict_words={}
    for i in range(len(list_words)):
        composition=tuple(sorted(tuple(list_words[i])))
        if composition not in dict_words.keys():
            dict_words[composition]=[]
        dict_words[composition].append(list_words[i])
    return dict_words
dict_words=anagram_finder(list_words)
dict_words

#the code creates an alphabetically ordered tuple out of each word in the list. If the tuple is 
#already present as a key in the dictionary dict_words, the new word with the same composition is 
#added to the list belonging to the key. If the tuple isn't present in the dictionary's keys, the 
#key is added to the dictionary and the word is added as a value. 
```

```python
#only the lists of anagrams with more than 5 elements are added to the list of anagrams that is
#printed on the screen 
anagrams_six={key:dict_words[key] for key in dict_words if len(dict_words[key])>5}
print(len(anagrams_six))
list_anagrams=[anagrams_six[key] for key in anagrams_six]
list_anagrams
```
